```
#Construção de um jogo de Forca utilizando Python

#importação de bibliotecas
import random

#Declaração da Função principal para execução do jogo
lista_palavras = [
                  'banana',
                  'jaboticaba',
                  'mamao',
                  'caju',
                  'melancia',
                  'laranja',
                  'mangostin',
                  'uva',
                  'pitaya',
                  'caja',
                  'morango',
                  'atemoia',
                  'pera'
                  ]
def jogar():

#Chamada da função que imprime a mensagem inicial do nosso jogo

  mensagem_inicial('Forca')

#Declaração de variaveis globais necessárias para nosso jogo

  palavra_secreta = escolhe_palavra_secreta()
  enforcou = False
  acertou = False
  quantidade_maxima_erros = 7
  erros = 0

  '''A variavel acertos é declarada e inicializada a partir do retorno da função 
  que cria letras acertadas'''

  acertos = criar_letras_acertadas(palavra_secreta)

  '''função de loop/laço que dita a execução do nosso jogo onde a condição de 
  continuidade são as variaveis acerto e enforcou serem falsas'''
  
  while (not acertou and not enforcou):

    #Solicita um chute de alguma letra para o jogador

    chute = input('Qual letra? {}\n'.format(acertos))

    '''Inicializada a variavel index para sabermos qual posição da nossa palavra
    estamos durante o loop `for`'''

    index = 0

    '''Nesse ponto a partir de um laço comparamos a letra chutada pelo jogador
    com todas as letras da nossa palavra secreta para verificar se temos um
    acerto e a posição que esse acerto se encontra na nossa palavra'''

    encontrou_letra = False

    for letra in palavra_secreta:
      if (chute.upper() == letra.upper()):
        encontrou_letra = True
        print('Encontrei a letra {} na posição {}'.format(chute, index + 1))
        acertos[index] = letra
      index += 1 #Lembrando que aqui precisamos incrementar o nosso index :)



    '''Nesse if condicional validamos se o jogar já acertou todas as letras
    para alterarmos o valor da variavel acertou e finalizarmos o nosso loop'''
    
    if ('_' not in acertos):
      print('voce ganhou!!')
      acertou = True
    elif (not encontrou_letra):
      erros += 1
      enforcou = erros == quantidade_maxima_erros


  if (enforcou):
    print('você não descobriu a palavra secreta que era: {}'.format(palavra_secreta))

  print('Fim do jogo!')

# Funcão que imprime a mensagem inicial do nosso jogo

def mensagem_inicial(nome_jogo):
  print('****************************')
  print('*Bem vindo ao jogo {}*'.format(nome_jogo))
  print('****************************')

#Função que cria lista de letras para validarmos os acertos

def criar_letras_acertadas(palavra):
  letras_acertadas = []
  for caractere in palavra:
    letras_acertadas.append('_')
  return letras_acertadas

def escolhe_palavra_secreta ():
  palavra = random.choice(lista_palavras)
  return palavra
```

```
# Criar uma função que receba uma lista de numeros e retorne eles ordenados, 
# caso a lista seja vazia retornar uma lista vazia.

# [1,2,3,5,4,10,8] = []
# []

# retorne:

# [1,2,3,4,5,8,10] = sorted_list

# []

# 5,4,3,2,1
# 1 execução:
# menor = 1
# lista sortedada = [1]
# lista original = 5,4,3,2
# 2 execução
# menor = 2
# lista sorteada = [1,2]
# lista original = 5,4,3


def ordena_numeros(numeros):
  data_list = numeros
  sorted_list = []

  while (len(data_list) > 0):
    menor_numero = data_list[0] # numero arbitrario da minha lita original
       
    for num in data_list:
      if (num < menor_numero):
        menor_numero = num
    
    sorted_list.append(menor_numero)
    data_list.remove(menor_numero)
    
  return sorted_list

qualquer_coisa = [32,8,55,00,14,43,30,16]

ordena_numeros(qualquer_coisa)
```

```
def tableDataText(table):       
    rows = []
    trs = table.find_all('tr')
    headerow = [td.get_text(strip=True) for td in trs[2].find_all('th')] # header row
    if headerow: # if there is a header row include first
        rows.append(headerow)
        trs = trs[2:]
    for tr in trs: # for every table row
        rows.append([td.get_text(strip=True) for td in tr.find_all(['th','td'])]) # data row
    return rows
```

```
list_table = tableDataText(tabela)
```

```
import pandas as pd
dftable = pd.DataFrame(list_table)
headers = dftable.iloc[0]
#dftable = dftable.rename(columns=dftable.iloc[0])
dftable  = pd.DataFrame(dftable.values[2:12], columns=headers)
dftable
```
